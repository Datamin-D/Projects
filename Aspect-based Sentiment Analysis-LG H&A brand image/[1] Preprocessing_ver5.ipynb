```python
#ver3,4용
import pandas as pd
import os

os.chdir(r'C:\Users\MinSeong\Desktop\LG 플젝\코드 모음\[1]preprocessing')
df = pd.read_csv("reviews_before_proprocessing_ver1.csv",index_col = 0)
df
```

```python
# title , message 분리하는 코드
title = df.drop('Message',axis = 1)
message = df.drop('Title',axis = 1)
title["Message"] = title["Title"]

mention = message.append(title)
mention = mention.drop("Title",axis = 1)
mention
```

```python
mention = mention.dropna(subset=["Message"]) #Message가 결측치면 삭제
mention.reset_index(drop = True, inplace = True)
df = mention
df['contents'] = df["Message"]
df = df.drop("Message",axis = 1)
df
```

```python
# # title, message 합치기. --> 안함. 문제 있는 코드. - 멘션에서 특히.
# df['contents'] = df['Title'] + " " + df['Message']
# df['contents']
```

```python
# 
from tqdm import tqdm
tqdm.pandas()

import nltk

df['contents'] = df['contents'].apply(lambda x: str(x)) #str 아닌 에러.
df['sents'] = df['contents'].progress_apply(lambda sents: nltk.sent_tokenize(sents))
df
```

```python
# 중간 저장
# drop_col = ['Title','Message','contents']
drop_col = ['contents']
df = df.drop(drop_col, axis = 1)
df.to_csv('sent_df.csv')
```

```python
# 리스트 형태에 맞게 변환

def remove_link_punc(customer_sents):
    docu = []
    for sent in customer_sents:
        # removing links # 
#         temp_sent = re.sub('http[s]?://(?:[a-zA-Z]|[0–9]|[$-_@.&+]|(?:%[0–9a-fA-F][0–9a-fA-F]))+', ' ', sent) #기존코드
#         temp_sent = re.sub('http[s]?:.* ', ' ', sent) #추가코드
        temp_sent = re.sub('http[s]?.*', ' ', sent)
        
        # removing all everything except a-z english letters # 작동
        regex = re.compile('[^a-zA-Z]')
        temp_sent = regex.sub(' ', temp_sent)

        # removing extra spaces 
        clean_sent = re.sub(' +', ' ', temp_sent).lower()
        docu.append(clean_sent)
    return docu
```

```python
from nltk.stem import WordNetLemmatizer

import nltk
# nltk.download('wordnet')
# nltk.download('stopwords')
from nltk.corpus import stopwords
from tqdm import tqdm
tqdm.pandas()



stop = stopwords.words('english')


def preprocessing(customer_sents):   #Preprocessing
    docu = []
    for sent in customer_sents:
        # tokenize into words
        tokens = [word for word in nltk.word_tokenize(sent)] 

        # remove stopwords
    #     stop = stopwords.words('english')
        tokens = [token for token in tokens if token not in stop]

        # remove words less than three letters
        tokens = [word for word in tokens if len(word) >= 3]

        # lower capitalization
        tokens = [word.lower() for word in tokens]

        # lemmatize
        lmtzr = WordNetLemmatizer()
        tokens = [lmtzr.lemmatize(word) for word in tokens]
        preprocessed_text = tokens
#         preprocessed_text= ' '.join(tokens)
        docu.append(preprocessed_text)
    return docu

df['ae_pre']=df['remove_link_punc'].progress_apply(lambda customer_sents: preprocessing(customer_sents))
```

```python
import re
from tqdm import tqdm
tqdm.pandas()

def remove_link_punc_unicode(string):
  # removing links # 정상 작동
  temp_string = re.sub('http[s]?://(?:[a-zA-Z]|[0–9]|[$-_@.&+]|(?:%[0–9a-fA-F][0–9a-fA-F]))+', ' ', string)
  
  # get lost \n \t \v 
  temp_string = re.sub("[\t\n\r\f\v]"," ",temp_string)

  # get lost unicode  
  temp_string = temp_string.encode("ascii", "ignore").decode('utf-8')
  # removing all everything except a-z english letters # 작동
  regex = re.compile('[^a-zA-Z]')
  temp_string = regex.sub(' ', temp_string)

#   removing extra spaces #작동
  clean_string = re.sub(' +', ' ', temp_string).lower()

  return clean_string
  

review_sents['remove_link_punc'] = review_sents['sent'].progress_apply(lambda customer_sents: remove_link_punc_unicode(customer_sents)) 
```

### tokenization, preprocessing

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
from tqdm import tqdm
tqdm.pandas()



stop = stopwords.words('english')


def preprocessing(text):   #Preprocessing
    # tokenize into words
    tokens = [word for sent in nltk.sent_tokenize(text) for word in nltk.word_tokenize(sent)] 
    
#     remove stopwords
    stop = stopwords.words('english')
    tokens = [token for token in tokens if token not in stop]
#     print(tokens)
    
    # remove words less than three letters
    tokens = [word for word in tokens if len(word) >= 3]
#     print(tokens)
    # lower capitalization
    tokens = [word.lower() for word in tokens]
#     print(tokens)
    # lemmatize
    lmtzr = WordNetLemmatizer()
    tokens = [lmtzr.lemmatize(word) for word in tokens]
#     print(tokens)
    preprocessed_text = tokens
#     preprocessed_text= ' '.join(tokens)

    # Stemming
#     stemmer = PorterStemmer()
#     preprocessed_text = [stemmer.stem(w) for w in preprocessed_text]
    
    return preprocessed_text

review_sents['pre_sent'] = review_sents['remove_link_punc'].progress_apply(lambda customer_sents: preprocessing(customer_sents)) 
```
